#JOINING DATA FRAME

```python
import pandas as pd

offers=pd.read_csv("/content/drive/MyDrive/Data Analytics/p01/data_offers.csv")
offers.head()
```

```python
#merging the orders and offer table based on order_gk using inner join
df=pd.merge(left=orders, right=offers,how='inner',on='order_gk')
#df.head()
df.shape
```

#Transform Data


```python
#replace 1 with yes and 0 with no in is_driver_assigned_key
df.head()
```

```python
import numpy as np
#np.where(df["is_driver_assigned_key"]==1,"yes","no") # it will create list of yes and no
value=np.where(df["is_driver_assigned_key"]==1,"yes","no")
# we need to make new column with slightly different name and assign the list
df["is_driver_assigned"]=value
df.head()
```

```python
#update order_status_key
val=np.where(df["order_status_key"]==4,"cancelled","system reject")
df["order_status"]=val
df.sample(10)
```

```python
#drop both old columns is_driver_assigned_key and order_status_key
df.drop(columns=["order_status_key","is_driver_assigned_key"],inplace=True)
# or simply reassign
#df=df.drop(columns=["order_status_key","is_driver_assigned_key"])
df.head()
```

```python
#rename the column
df=df.rename(columns={"order_datetime":"order_time"})
df.head()
```

#GROUPING DATA

```python
#group the order status and count their orger gk means order id
df.groupby(["order_status"])['order_gk'].count()
```

```python
#want to group more column
df.groupby(['is_driver_assigned','order_status'])['order_gk'].count()
```

```python
df.groupby(["is_driver_assigned"])['order_gk'].count()
```

#Pivot Table


```python
#extract the information using the pivot table
pivot=df.pivot_table (columns =["is_driver_assigned" , "order_status"],
               values="order_gk",
               aggfunc="count")
df.head()
```

```python
#plot the pivot table
pivot.plot(kind="bar")
```

```python
#df.sample(4)
#ectraction of order failure w.r.t time in hours
datetime=df.loc[0,'order_time']
#str(datetime).split(":") to get only hours
str(datetime).split(":")[0]
```

```python
#now if we want to calculate time for all rows use apply function
df['order_hour']=df['order_time'].apply(lambda x : str (x).split(":")[0])
df.sample()
```
